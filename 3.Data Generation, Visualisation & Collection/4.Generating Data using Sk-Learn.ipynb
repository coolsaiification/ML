# Generating datasets
functions available in sklearn.datasets
- make_regression
- make_classification
- make_blobs
- make_moons

```python
from sklearn.datasets import make_regression, make_classification, make_moons, make_blobs
```

```python
import matplotlib.pyplot as plt
```

## Regression

```python
XR,YR = make_regression?
```

```python
XR,YR = make_regression(n_samples=200,n_features=1,n_informative=1,noise=15.2)
print(XR.shape, YR.shape)
```

```python
plt.scatter(XR,YR)
plt.show()
```

## Classification

```python
XC,YC = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=5)
# Random state for saving data
print(XC.shape, YC.shape)
```

```python
plt.scatter(XC[:,0], XC[:,1], c=YC)
plt.show()
```

## Blobs Dataset for Clustering

```python
XB, YB = make_blobs?
```

```python
XB, YB = make_blobs
```

```python
XB, YB = make_blobs(n_samples=1000, n_features=2, centers=5)
print(XB.shape, YB.shape)
```

```python
plt.scatter(XB[:,0], XB[:,1], c=YB)
plt.show()
```

## Moons Data
requires non linear solution

```python
XM, YM = make_moons(n_samples=100, noise=0.1)
print(XM.shape, YM.shape)
```

```python
plt.scatter(XM[:,0], XM[:,1], c=YM)
plt.show()
```

## Subplots

```python
plt.figure(figsize=(10,10))

plt.subplot(221)
plt.title("Regression")
plt.scatter(XR,YR)

plt.subplot(222)
plt.title("Classification")
plt.scatter(XC[:,0], XC[:,1], c=YC)

plt.subplot(223)
plt.title("Blobs")
plt.scatter(XB[:,0], XB[:,1], c=YB)

plt.subplot(224)
plt.title("Moons")
plt.scatter(XM[:,0], XM[:,1], c=YM)

plt.show()
```
