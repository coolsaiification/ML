# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(10)
x **= 2
x2 = x**3
print(x)
```

```python
plt.plot(x)
plt.show()
plt.plot(x2)
plt.show()
# Together
plt.plot(x, color='red', label='2019')
plt.plot(x2, color='cyan', label='2018')
plt.legend()
plt.show()
```

```python
# Plot points
x = np.arange(10)
y = x*2
# Triange marker instead of circle
plt.scatter(x,y, color='purple', label='Stock Price', marker='^')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Price Prediction")
plt.show()
```

## Bar Chart

```python
numbers = np.random.randint(1,10,5)
numbers2 = np.random.randint(1,10,5)
indices = np.arange(5)
print(numbers)
print(numbers2)
print(indices)
```

```python
plt.bar(indices, numbers, 0.25, color='yellow', label='Gold')
plt.bar(indices+0.25, numbers2, 0.25, color='red', label='Red')
plt.legend()
plt.plot()
```

## Pie Chart

```python
labels = ['Calculus', 'Linear Algebra', 'Python', 'Algorithms']
values = [80,94,30,150]
plt.pie(values, labels=labels, shadow=True, radius=1.5, explode=(1,0,0,0))
plt.plot()
```

## Histogram 

```python
# Generate Normal Distribution 1-D

# For Standard Normal Distribution u=0, sigma=1
u = 5
sigma = 2

vals = u + sigma*np.random.randn(300)
print(vals.shape)
```

```python
# No. of intervals = 50
plt.hist(vals,50)
plt.show()
```

```python
x = vals
y = np.zeros(300)
# Most points near 5
plt.scatter(x,y)
plt.show()
# In normal distribution >95% points in range u-2*sigma to u+2*sigma
```

```python
vals = np.round(vals)
z = np.unique(vals, return_counts=True)
print(z)
```

```python
# Multivariate Normal Distribution

# Bivariate
mean1 = np.array([0,0])
covar1 = np.array([[1,0.8],[0.8,1]])
dist1 = np.random.multivariate_normal(mean1, covar1, 500)
print(dist1.shape)

mean2 = np.array([5,6])
covar2 = np.array([[1.2,-0.5],[-0.5,1.3]])
dist2 = np.random.multivariate_normal(mean2, covar2, 500)
print(dist2.shape)
```

```python
plt.scatter(dist1[:,0], dist1[:,1])
plt.scatter(dist2[:,0], dist2[:,1])
plt.show()
```
