# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(10)
x **= 2
x2 = x**3
print(x)
```

```python
plt.plot(x)
plt.show()
plt.plot(x2)
plt.show()
# Together
plt.plot(x, color='red')
plt.plot(x2, color='cyan')
plt.show()
```

```python
# Plot points
x = np.arange(10)
y = x*2
# Triange marker instead of circle
plt.scatter(x,y, color='purple', label='Stock Price', marker='^')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Price Prediction")
plt.show()
```
