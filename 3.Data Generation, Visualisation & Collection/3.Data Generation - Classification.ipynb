# Binary Classification

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Apple
# Means of 2 features (size and sweetness)
u1 = np.array([2,4])
# covariance xx,xy,xy,yy
# xy +ve -> Both increase together   (+ve slope)
cov1 = np.array([[2,-0.8],
                [-0.8,1]])
# Multivariate normal as there are multiple features
X1 = np.random.multivariate_normal(u1, cov1, 300)
print(X1.shape)
```

```python
# Orange
u2 = np.array([-1,-3])
cov2 = np.array([[1,0],
                [0,2]])
X2 = np.random.multivariate_normal(u2, cov2, 400)
```

```python
plt.scatter(X1[:, 0], X1[:, 1], label='Apple')
plt.scatter(X2[:, 0], X2[:, 1], marker='^', label='Orange')
plt.legend()
plt.title('Apples and Oranges')
plt.xlabel('x1(Size)')
plt.ylabel('x2(Sweetness)')
plt.show()
```

## Generating data

```python
Y = np.zeros(700)
# 1 for apples, 0 for oranges
Y[:300]=1
print(Y.shape)

# Combine X1 and X2 into X with 700 ( 300+400) rows
X = np.vstack((X1,X2))
print(X.shape)
```

```python
# Without labels
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# With labels
for i in range(len(Y)):
    if(Y[i]==1):
        plt.scatter(X[i][0], X[i][1], color='Blue')
    else:
        plt.scatter(X[i][0], X[i][1], color='Orange', marker='^')
plt.show()
```

```python
#Shortcut
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```
