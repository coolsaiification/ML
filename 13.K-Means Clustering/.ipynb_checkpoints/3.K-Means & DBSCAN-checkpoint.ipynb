## Problem with K-Means
It can fail with complex datasets

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles, make_moons
```

```python
X,Y = make_moons(n_samples=200, noise=0.1)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
```

```python
km.fit(X)
```

```python
centers = km.cluster_centers_
labels = km.labels_

plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centers[:,0], centers[:,1], color='red', marker='*', s=200)
plt.show()
```

## DBSCAN
- Density Based Spatial Clustering of Applications with Noise
- Based upon on the idea, that a __cluster is a high density area surrounded by low density region__
- Starts by exploring a small area, if density is "good enough" it considered as part of cluster, and explores neigbours and so on, so it groups regions based upon density!

```python
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
DBSCAN?
```

```python
# eps = threshold minimum distance
# min_samples = nbd pts
dbs = DBSCAN(eps=0.2, min_samples=5)
```

```python
yPred = dbs.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=yPred)
plt.show()
```
