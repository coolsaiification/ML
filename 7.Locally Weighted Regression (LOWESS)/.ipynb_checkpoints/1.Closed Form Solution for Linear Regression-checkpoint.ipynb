# theta = inverse(transpose(X)X) transpose(X)Y

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# Generate dataset

x,y = make_regression(n_samples=400, n_features=1, n_informative=1, noise=1.8, random_state=11)
y.reshape((-1,1))
print(x.shape, y.shape)
```

```python
# Normalize

x = (x-x.mean())/x.std()
```

```python
# Visualize

plt.scatter(x,y)
plt.title('Normalized Data')
plt.show()
```

```python
# Add ones to x in last column for theta0 
ones = np.ones((x.shape[0],1))
x_ = np.hstack((x,ones))
print(x_.shape)
```

```python
def predict(x,theta):
    return np.dot(x,theta)

def getThetaClosedForm(x,y):
    firstPart = np.dot(x.T,x)
    secondPart = np.dot(x.T,y)
    secondPart = np.reshape(secondPart, (-1,1))
    theta = np.dot(np.linalg.pinv(firstPart),secondPart)
    theta = np.reshape(theta,(-1,1))
    return theta
```

```python
theta = getThetaClosedForm(x_,y)
print(theta)
```

```python
# Visualize

plt.scatter(x,y)
plt.plot(x, predict(x_,theta), color='red', label='Prediction')
plt.title("Normalized Data with Prediction")
plt.legend()
plt.show()
```
