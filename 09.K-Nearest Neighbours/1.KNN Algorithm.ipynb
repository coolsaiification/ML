```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = pd.read_csv("../Datasets/xdata.csv").values
y = pd.read_csv("../Datasets/ydata.csv").values

x = x[:,1:]
y = y[:,1:].reshape((-1,))

print(x.shape, y.shape)
```

```python
query_x = np.array([2,3])
plt.scatter(x[:,0], x[:,1], c=y)
plt.scatter(query_x[0], query_x[1], color='red', marker='^')
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(x,y,queryPoint,k=5):
    val = []
    for i in range(x.shape[0]):
        val.append((dist(queryPoint,x[i]),y[i]))
    # Sorting based on distance(first value of tuple)
    val = sorted(val)
    # First/Nearest k points
    val = val[:k]
    val = np.array(val)
    new_vals = np.unique(val[:,1], return_index=True)
    
    max_freq_index = new_vals[1].argmax()
    prediction = new_vals[0][max_freq_index]
    
    return prediction
```

```python
print(knn(x,y,[4,5]))
```

```python
print(knn(x,y,[0,0]))
```
