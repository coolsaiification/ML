```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = pd.read_csv('../Datasets/Locally Weighted Regression/weightedX.csv').values
y = pd.read_csv('../Datasets/Locally Weighted Regression/weightedY.csv').values

X = (x-x.mean())/x.std()
Y = y
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

## Object

```python
model = LinearRegression()
```

## Training

```python
model.fit(X,Y)
```

## Predictions

```python
output = model.predict(X)
```

## Parameters Learnt

```python
bias = model.intercept_
coeff = model.coef_

print(bias)
print(coeff)
```

## Score

```python
print(model.score(X,Y))
```

## Visualize

```python
plt.scatter(X,Y,label='Data')
plt.plot(X,output,color='red',label='Prediction')
plt.legend()
plt.show()
```

# Not a linear relationship

```python
x1 = X
x1 = np.reshape(x1,(-1,))
x2 = X**2
x2 = np.reshape(x2,(-1,))
print(x1.shape,x2.shape)
XN = np.stack((x1,x2),axis=1)
print(XN.shape)
```

```python
model2 = LinearRegression()
model2.fit(XN,Y)
output2 = model2.predict(XN)

print(model2.intercept_)
print(model2.coef_)
print(model2.score(XN,Y))
```

```python
# Visualize
plt.scatter(XN[:,0],Y,label='Data')
plt.scatter(XN[:,0],output2,label='Prediction',color='red')
plt.legend()
plt.show()
```
